```python
import time
import random
import string

def generate_random_text():
    text_length = random.randint(50, 200)
    random_text = "".join(random.choice(string.ascii_letters + string.digits + string.punctuation) for i in range(text_length))
    return random_text

def get_user_input():
    text = generate_random_text()
    print("Practice typing the following text:")
    print(text)
    start_time = time.time()
    user_input = input()
    end_time = time.time()
    elapsed_time = end_time - start_time
    return text, user_input, elapsed_time

def calculate_speed_and_accuracy(text, user_input, elapsed_time):
    correct_characters = sum(c1 == c2 for c1, c2 in zip(text, user_input))
    total_characters = len(text)
    typing_speed = correct_characters / elapsed_time if elapsed_time > 0 else 0
    accuracy = correct_characters / total_characters if total_characters > 0 else 0
    return typing_speed, accuracy

def display_results(typing_speed, accuracy):
    print("Typing speed: %.2f WPM" % typing_speed)
    print("Accuracy: %.2f%%" % (accuracy * 100))

def main():
    while True:
        text, user_input, elapsed_time = get_user_input()
        typing_speed, accuracy = calculate_speed_and_accuracy(text, user_input, elapsed_time)
        display_results(typing_speed, accuracy)

        next_test = input("Do you want to practice typing again? (yes/no): ")
        if next_test == "no":
            break

if __name__ == "__main__":
    main()
```
